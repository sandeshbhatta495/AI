Standard deviation is a measure of how spread out a set of data is from its average (mean). A low standard deviation means most of the data points are clustered around the mean, while a high standard deviation indicates the data points are more widely dispersed. It is calculated by taking the square root of the variance, which is the average of the squared differences from the mean. 
What it tells you
Clustering: A low standard deviation indicates data points are close to the mean.
Spread: A high standard deviation indicates data points are spread far from the mean.
Consistency: It's used to compare the consistency of different data sets.
Variability: It quantifies the average amount of variability in your data set. 
How it's calculated
Find the mean of the data set.
Calculate the variance:
Subtract the mean from each data point and square the result.
Find the average of these squared differences.
Find the standard deviation: Take the square root of the variance. 
Why use standard deviation?
It's more useful than variance in real-world contexts because it's expressed in the same units as the original data, unlike variance which is in squared units.
It helps in statistical analysis, such as calculating statistical intervals and hypothesis tests.
