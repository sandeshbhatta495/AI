```python
# libraries 
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
X = data.data                # shape (n_samples, 30)
y = data.target              # 0 = malignant, 1 = benign
feature_names = data.feature_names
```

```python
# Create DataFrame for easy inspection
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
```

```python
# Quick checks
print("Shape:", df.shape)
print("\nFeature names (30):")
print(feature_names)
```

```python
df.head(5)
```

```python
# Descriptive statistics (mean, std, min, max)
display(df.describe().T)
```

```python
display(df.describe()) 
```

```python
# How many samples of each class?
print("\nClass distribution:")
print(df['target'].value_counts())   # 0 malignant, 1 benign
```

```python
# compare distribution of one important feature by class
import matplotlib.pyplot as plt
# plt.figure(figsize=(6,4))
plt.hist(df[df['target']==0]['mean radius'], bins=20, alpha=0.6, label='malignant')
plt.hist(df[df['target']==1]['mean radius'], bins=20, alpha=0.6, label='benign')
plt.xlabel('mean radius')
plt.legend()
plt.show()
```
