# The explanation of column or title 
For each nucleus, these are measured by an imaging algorithm:

radius — average distance from center to points on the perimeter (roughly size).

texture — measure of grayscale variation (how rough/smooth inside nucleus).

perimeter — length around the nucleus (correlates with size/shape irregularity).

area — area inside nucleus boundary (related to radius but not identical).

smoothness — local variation in radius (how smooth the border is).

compactness — 
high values → irregular shape.

concavity — severity of concave portions of the contour (indentations).

concave points — number of concave portions of the contour (count-like).

symmetry — how symmetric the shape is.

fractal_dimension — complexity of the boundary (0–2 type measure; higher → more complex/irregular).

Intuition: malignant tumors often have larger, more irregular nuclei (higher radius/area/perimeter), more concavities, less smoothness, and higher fractal dimension.
