```python
y_mean = y. mean()
```

```python
SSxy = (df.x_diff * df.y_diff).sum()
```

# for slope 

```python
m= SSxy/ SSxx
```

```python
# Calculation of intercept 
b = y_mean - m*x_mean
```

# Prdict The Value With Model 

```python
def predict(value):
    predict = m*value + b
    return predict
```
