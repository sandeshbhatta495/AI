```python
X,y = make_regression(n_samples =200 , n_features = 1, n_informative =1 , n_targets=1 , noise = 20)
```

```python
# finding the slope and intercept
m= lr.coef_
b=lr.intercept_
```

```python
class GDRegressor:
    def __init__(self, learning_rate, epochs):
        self.m = 98.08   # initial slope
        self.b = 120     # initial intercept
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self, X,y):
      #claculate the b using Gd
      loss_slope = -2*np.sum(y - self.m* X.ravel()- self.b)
      self.b=self.b - self.lr*loss_slope
      print(self.b)

# Re-instantiate gd with the corrected class definition
gd = GDRegressor(0.0001, 1000)
gd.fit(X,y)
```
