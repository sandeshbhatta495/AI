```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4, n_features=1 , n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

# using ordinary least sqaure method to find the the coff or it also called slope 

```python
#OLs 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

# its shows that the the slope is 78.35 also its called coef 

```python
# also finding intercept (b
reg.intercept_
```

```python
# now how our regression like look like 
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color ='red')
```

 # Its Shows that the value of b is find out by using standrad library and now we were trying without doing it .

```python
# step 1 initilization b to the random value let b=0 and slope is constant  m =78.35
y_pred =((78.35 * X ) + 0).reshape(4) 
# calculation of y prediction (y=mx+b)
y_pred
```

```python
plt.scatter(X, y)

# OLS model (with intercept)
plt.plot(X, reg.predict(X), color='red', label='by using OLS')
# for b=0
plt.plot(X,y_pred,color = 'blue',label='b = 0')

plt.legend()
plt.show()
```

```python
 m= 78.35
```

```python
b= 0
```

# Calculation of Loss Tangent 

```python
loss_slope = -2 * np.sum(y - m * X.ravel() - b)

# Mathmatical Equation 
# dl/db= -2 sumxetion(yi-mxi-b) where b=0 (loss_slope)
 # loss_Tangent (L) = sumxetion(i=0 to n)[yi-mxi-b]^2 
loss_slope
```

```python
lr =0.1
step_size = loss_slope*lr
step_size 
# mathmatical realtion
 # bnew=bold-itta*slope; and stepsize = itta*slope 
# lr =learning rate denoted by itta 
```

```python
# new b 
# calculation of new intercept 
b = b - step_size 
b
```

```python
# Calcualtion of new y_pred
y_pred1 =((78.35 * X ) + b).reshape(4) 
# calculation of y prediction (y=mx+b)
y_pred1
```

```python
plt.scatter(X, y)
# OLS model (with intercept)
plt.plot(X, reg.predict(X), color='red', label='by using OLS')
# for b=0
plt.plot(X,y_pred,color = 'blue',label='b = 0')
# for new b 
plt.plot(X,y_pred1,color = 'green',label='b = {}'.format(b))
plt.legend()
plt.show()
```
