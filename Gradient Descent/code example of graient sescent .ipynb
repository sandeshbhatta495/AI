```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=4, n_features=1 , n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

# using ordinary least sqaure method to find the the coff or it also called slope 

```python
#OLs 
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

# its shows that the the slope is 78.35 also its called coef 

```python
# also finding intercept (b
reg.intercept_
```

```python
# now how our regression like look like 
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color ='red')
```

```python
plt.scatter(X, y)

# OLS model (with intercept)
plt.plot(X, reg.predict(X), color='red', label='by using OLS')

# Forced b = 0 model
reg_no_intercept = LinearRegression(fit_intercept=False)
reg_no_intercept.fit(X, y)
plt.plot(X, reg_no_intercept.predict(X), color='blue', label='by using b = 0')

plt.legend()
plt.show()
```

```python
# by using manual calculation
def y_pred(x):
    return reg_no_intercept.coef_[0] * x
plt.plot(X, y_pred(X), color='blue', label='by using b = 0')
```
