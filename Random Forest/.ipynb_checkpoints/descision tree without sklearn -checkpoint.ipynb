```python
import random
from collections import Counter

# Gini impurity
def gini(y):
    counts = Counter(y)
    impurity = 1
    for c in counts.values():
        p = c / len(y)
        impurity -= p ** 2
    return impurity
```

```python
def split_data(X, y, feature, value):
    left_X, left_y, right_X, right_y = [], [], [], []
    for i in range(len(X)):
        if X[i][feature] <= value:
            left_X.append(X[i])
            left_y.append(y[i])
        else:
            right_X.append(X[i])
            right_y.append(y[i])
    return left_X, left_y, right_X, right_y
```
