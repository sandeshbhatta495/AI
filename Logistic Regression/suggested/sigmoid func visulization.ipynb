```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create input values from -10 to 10
x = np.linspace(-10, 10, 400)
```

```python
# Sigmoid function
def sigmoid(z):
    val = 1/(1 + np.exp(-z)
    return val
```

```python
# Compute sigmoid output
y = sigmoid(x)
```

```python
# Plot
plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```
