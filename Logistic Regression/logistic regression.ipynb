 ## Logistic Regression is a supervised learning method for classification (usually binary). It models the probability that an input belongs to class 1 by passing a linear combination of features through the sigmoid function (which squashes values to [0,1]). We train the model by maximizing the probability of the observed labels (equivalently minimizing log-loss / cross-entropy) and use the resulting probability with a threshold (commonly 0.5) to decide class labels.
