```python
import numpy as np
```

```python
# Training data
X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])
```

```python
# Initialize parameters
w = 0.0
b = 0.0
learning_rate = 0.01
epochs = 50
```

```python
# SGD algorithm
for epoch in range(epochs):
    for i in range(len(X)):
        y_pred = w * X[i] + b
        error = y_pred - y[i]

        # Compute gradients
        dw = 2 * error * X[i]
        db = 2 * error

        # Update parameters
        w = w - learning_rate * dw
        b = b - learning_rate * db
```
