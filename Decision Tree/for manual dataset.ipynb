```python
# Dataset: [Study, Attendance, Pass]
data = [
    ["High", "High", "Yes"],
    ["Low",  "High", "No"],
    ["High", "Low",  "Yes"],
    ["Low",  "Low",  "No"]
]

features = ["Study", "Attendance"]
```

```python
# Entropy function
import math

def entropy(data):
    label_count = {}
    total = len(data)

    # Count Yes / No
    for row in data:
        label = row[-1]
        label_count[label] = label_count.get(label, 0) + 1

    ent = 0
    for count in label_count.values():
        p = count / total
        ent -= p * math.log2(p)

    return ent
```

```python
# Information Gain Function
def information_gain(data, feature_index):
    parent_entropy = entropy(data)
    subsets = {}

    # Split data by feature values
    for row in data:
        key = row[feature_index]
        if key not in subsets:
            subsets[key] = []
        subsets[key].append(row)

    weighted_entropy = 0
    for subset in subsets.values():
        weighted_entropy += (len(subset) / len(data)) * entropy(subset)

    return parent_entropy - weighted_entropy
```
