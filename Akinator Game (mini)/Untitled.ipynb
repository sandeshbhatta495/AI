```python
import pandas as pd
import random

random.seed(42)

characters = ["dipi","sandi","rohit","sompal","kc","anil","kushal","bhutel"]


    

attributes = [
    "","fictional","male","female","movie","tv","anime","sports",
    "politics","music","scientist","alive","superpower","villain","animated"
]

data = []

for name in characters:
    row = {"character": name}
    for attr in attributes:
        row[attr] = random.randint(0, 1)
    data.append(row)

df = pd.DataFrame(data)
df.to_csv("akinator_data.csv", index=False)

print("CSV file created with shape:", df.shape)
```
